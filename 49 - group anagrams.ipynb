```python
# 49 - Group Anagrams
"""
    Given an array of strings strs, group the anagrams together. You can return the answer in any order.

    Example 1:
        Input: strs = ["eat","tea","tan","ate","nat","bat"]
        Output: [["bat"],["nat","tan"],["ate","eat","tea"]]

    Example 2:
        Input: strs = [""]
        Output: [[""]]

    Example 3:
        Input: strs = ["a"]
        Output: [["a"]]
"""
```

```python
# My Accepted solution - but Time Limit Exceeded!
#
class Solution(object):
    def groupAnagrams(self, strs):
        """
        :type strs: List[str]
        :rtype: List[List[str]]
        """
        def str2dict(str):
            dict = {}
            for char in str:
                if char not in dict:
                    dict[char] = 1
                else:
                    dict[char] += 1
            return dict

        def isAnagrams(s, t):
            return str2dict(s) == str2dict(t)

        output = []
        if len(strs) <= 1:
            return [strs]
        while(strs):
            ref = strs[0]
            ana = []
            remaining = []
            for string in strs:
                if isAnagrams(string, ref):
                    ana.append(string)
                else:
                    remaining.append(string)
            output.append(ana)
            strs = remaining
        
        return output
```

```python
# A better solution
#
class Solution(object):
    def groupAnagrams(self, strs):
        """
        :type strs: List[str]
        :rtype: List[List[str]]
        """
        from collections import defaultdict
        groups = defaultdict(list)

        for s in strs:
            key = "".join(sorted(s))
            groups[key].append(s)
        
        return list(groups.values())
    
# Time complexity: O(n*k*log(k)), while n = len(strs), k = max(len(str)) for str in strs
# Space complexity: O(n*k) 
```
