```python
# 202 - Happy Number
"""
    Write an algorithm to determine if a number n is happy.

    A happy number is a number defined by the following process:

    Starting with any positive integer, replace the number by the sum of the squares of its digits.
    Repeat the process until the number equals 1 (where it will stay), or it loops endlessly in a cycle which does not include 1.
    Those numbers for which this process ends in 1 are happy.
    Return true if n is a happy number, and false if not.

    Example 1:
        Input: n = 19
        Output: true
        Explanation:
        1^2 + 9^2 = 82
        8^2 + 2^2 = 68
        6^2 + 8^2 = 100
        1^2 + 0^2 + 0^2 = 1

    Example 2:
        Input: n = 2
        Output: false
"""
```

```python
# My accepted solution:
#
class Solution(object):
    def isHappy(self, n):
        """
        :type n: int
        :rtype: bool
        """
        def iterate(num):
            temp = []
            new = 0
            for char in str(num):
                temp.append(int(char)) 
            for i in range(len(temp)):
                new += temp[i]**2
            return new

        results = {}
        while n not in results:
            results[n] = 1
            n = iterate(n)
            if n == 1:
                return True
        return False

# Time complexity: let d = number of digits of n
#                  for iterate() -> O(d)
#                  for happy number loop run -> O(1), because the number of iterations is bounded by a constant, independent of the input size
#                  so time complexity = O(d) = O(log(n))
# Space complexity: temp -> O(d), results -> O(1), so space complexity = O(d) = O(log(n)) 
```
