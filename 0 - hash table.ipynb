```python
class HashTable:

    def __init__(self, size):
        self.size = size
        self.hash_table = [[] for _ in range(self.size)]

    def _hash(self, key): # calculate index
        return hash(key) % self.size
    
    def insert(self, key, value):
        index = self._hash(key)
        bucket = self.hash_table[index]

        for i, (k, v) in enumerate(bucket):
            if k == key:
                bucket[i] = (key, value)
                return
        bucket.append((key, value))

    def get(self, key):
        index = self._hash(key)
        bucket = self.hash_table[index]
        
        for (k, v) in bucket:
            if k == key:
                return v
        raise KeyError(f"key '{key}' not found")

    def remove(self, key):
        index = self._hash(key)
        bucket = self.hash_table[index]

        for i, (k, v) in enumerate(bucket):
            if k == key:
                del(bucket[i])
                return
        raise KeyError(f"key '{key}' not found")

    def __str__(self): # print format of hash table
        result = []
        for i, bucket in enumerate(self.hash_table):
            result.append(f"{i}: {bucket}")
        return "\n".join(result)
```

```python
ht = HashTable(size=5)

ht.insert("apple", 10)
ht.insert("banana", 20)
ht.insert("grape", 30)
ht.insert("orange", 40)

print(ht.get("apple"))

ht.insert("apple", 99)
print(ht.get("apple"))

ht.remove("banana")

print("\ncurrent hash table:")
print(ht)
```
